## File Handling
The key function for working with files in Python is the open() function.

The open() function takes two parameters; filename, and mode.

There are four different methods (modes) for opening a file

**"r"** - Read - Default value. Opens a file for reading, error if the file does not exist


**"a"**- Append - Opens a file for appending

**"w"** - Write - Opens a file for writing, creates the file if it does not exist

**"x"** - Create - Creates the specified file, returns an error if the file exists

Writing to a File

```python
with open("example.txt", "w") as file:
    file.write ("Hello World")
```

## Reading CSV Files in Python

A CSV (or Comma Separated Value) file is the most common type of file that a data scientist will ever work with. These files use a “,” as a delimiter to separate the values and each row in a CSV file is a data record.


These are useful to transfer data from one application to another and is probably the reason why they are so commonplace in the world of data science.

```python
#The Pandas library makes it very easy to read CSV files using the read_csv() function:
```

```python

import csv
#Writing to a csv file
with open ("data.csv", "w", newline='') as file:
  writer = csv.writer(file)
  writer.writerow(["Name","Age","City"])
  writer.writerow(["John", "25", "New York"])
  writer.writerow(["Jane", "30", "London"])
```

```python
# reading from a csv file

with open  ("data.csv", "r") as file:
  reader = csv.reader(file)
  for row in reader:
    print(row)
```

### Working with JSON files

The json module helps to work with JSON (JavaScript Object Notation) format


```python
import json
#from json import dump,load
#writing dictionary to a json file
data = {"name": "John","age": 25, "city": "New York"}

with open("data.json","w") as file:
  json.dump(data,file)
```

```python
# Reading from a json file
with open("data.json","r") as file:
  data = json.load(file)
  print(data)
```

## Using statement for Resource Management


The with statement automatically closes files when done, this is considered best practice to avoid leaving files open , which oculd leads to resource leaks.

```python
with open("data.csv","r") as file:
  content = file.read()
  print(content)
  # the file is automatically closed after this block, no need to manually close it
```

```python
file = open("data.csv","r")
content = file.read()
file.close() #you must remember to close the file
```

## Importing and Using Python Modules in Python

### What Are Python Modules?
A module is a file containing Python definitions and statements (like functions, classes, and variables) that can be reused in other programs.
Python modules allow you to organize your code into smaller, more manageable files and reuse code across multiple projects.


## *Types of Modules in Python*
*Built-in Modules:*

Pre-installed modules that come with Python.

Examples: math, os, random, sys,

*External/Third-Party Modules:*

Modules installed via a package manager like pip.

Examples: numpy, pandas, matplotlib.

### Import an entire module:

```python
import math
print(math.sqrt(16))
```

### 2. Import With Alias

Use an alias to shorten the module name:

```python
import numpy as np
arr = np.array([1, 2, 3])
print(arr)
```

### 3. Import Specific Functions or Attributes

import only the required functions:

```python
from math import sqrt, pi
print(sqrt(16))
print(pi)
```

### 4. Import All Functions
Import everything from a module:

```python
from math import *
print(cos(0))
#Note: Avoid using this unless necessary, as it can cause naming conflicts.
```
